```python
import urllib.request, urllib.parse , urllib.error
from bs4 import BeautifulSoup
import ssl
```

```python
# ignore ssl certificate errors 
ctx = ssl.create_default_context()
ctx.check_hostname = False
ctx.verify_mode = ssl.CERT_NONE

url = input("Enter -")
html = urllib.request.urlopen(url, context=ctx).read() # read whole 
soup = BeautifulSoup(html, 'html.parser') # parse html

# retrieve all of anchor tags 
tags = soup('a') 
for tag in tags:
    print("Link :",tag.get('href', None)) # only href attribute

# http://dr-chuck.com 
```

```python
# ignore ssl certificate errors 
ctx = ssl.create_default_context()
ctx.check_hostname = False
ctx.verify_mode = ssl.CERT_NONE

url = input("Enter -")
html = urllib.request.urlopen(url, context=ctx).read() # read whole 
soup = BeautifulSoup(html, 'html.parser') # parse html

# retrieve all of anchor tags 
count = 0
sum_num = 0
tags = soup('span') 
for tag in tags:
    #print(tag) 
    num = int(tag.get_text())
    #print(type(num))
    count+=1
    sum_num+= num
    print(num)
    
print(count)
print(sum_num)
```
